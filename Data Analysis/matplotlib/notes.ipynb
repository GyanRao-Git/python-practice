```python
import matplotlib.pyplot as plt
```

### LINE PLOT

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
plt.plot(x,y)
plt.title("num vs square")
plt.xlabel("num")
plt.ylabel("squares")
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='--',marker='o',linewidth=3,markersize=9)
plt.grid(True)
plt.title("num vs square")
plt.xlabel("num")
plt.ylabel("squares")
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))
plt.subplot(1,2,1)
plt.plot(x,y1,color='green')

plt.subplot(1,2,2)
plt.plot(x,y2,color='blue')
 
```

### BAR PLOT

```python
cat=['A','B','C','D']
val=[3,7,2,1]

plt.bar(cat,val,color='blue')
```

```python
import pandas as pd
import numpy as np
```

```python
filename='../sales_data_sample.csv'
df=pd.read_csv(filename,encoding='latin1')
df.head()
```

```python
df.describe()
```

```python
x=df['COUNTRY']
y=df['QUANTITYORDERED']

plt.figure(figsize=(20,10))
plt.bar(x,y)
plt.yticks(np.arange(0, y.max() + 2, 10))
plt.grid(True)
plt.show()
```

```python
var=df.groupby('COUNTRY')['QUANTITYORDERED'].sum()
plt.figure(figsize=(15,10))
plt.pie(var,labels=var.index)
```
